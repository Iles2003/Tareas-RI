## 2. Limpieza de datos

```python
## se guarda el el corpus en una lista 
vocabulario = list(set(text.lower().replace('\n',' ').replace('.',' ').replace(',',' ').split()))
vocabulario
```

```python
## ver longitud de la variable vocabulario
N = len(vocabulario)
N
```

## 3. Generacion del corpus

```python
## aqui se divide el corpus en documentos en donde cada salto de linea o "\n" es un indicador para separar
corpus = text.split('\n')
corpus
```

```python
## en una caribale N se guarda el # de documentos total
len(corpus)
 
```

## Definimos una funcion para calcular el TF (Termine Frequecy)

```python
def get_tf(doc):
    vec = []
    palabras_doc = doc.lower().replace('.','').replace(',','').split()
    for word in vocabulario:
        cont = palabras_doc.count(word)
        vec.append(cont)
    return np.array(vec)
```

```python
# guardamos en una variable "x" el primer documento
x = get_tf(corpus[0])
x
```

```python
## se genera la matriz TF en donde para todos los documentos del corpus
## se guarda la lista creada con "get_tf(doc)" en un vector y luego se
## separa cada vector y se coloca otro despues de otro para crear una matriz
corpus_vec = []
for doc in corpus:
    doc_vec = get_tf(doc)
    corpus_vec.append(doc_vec)
```

```python
## espacio vectorial TF
print(corpus_vec)
```

```python
## se convierte la matrix TF a una tipo Numpy para optimizar operaciones
matriz_tf = np.array(corpus_vec)
matriz_tf
```
